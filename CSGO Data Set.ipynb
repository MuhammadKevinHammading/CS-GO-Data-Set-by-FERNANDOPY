# 1. CS:GO Players

```python
csgo_player = pd.read_csv('players.csv')
csgo_player.head()
```

```python
csgo_player.info()
```

Column 'ID' dtype are float, needs to be change it with int.\
There is null value.

```python
csgo_player.isna().sum()
```

Column 'ID' and 'NickName' contain a lot of missing values. while the column 'Name' has only 1 missing value.\
Fill missing values in column 'ID' with 0 then change dtype to int.

```python
csgo_player['ID'] = csgo_player['ID'].fillna(0)
csgo_player['ID'] = csgo_player['ID'].astype('int64')
```

```python
csgo_player.duplicated().sum()
```

There are duplicate values, need to be deleted.

```python
csgo_player.drop_duplicates(inplace=True)
csgo_player.reset_index(drop=True, inplace=True)
csgo_player.duplicated().sum()
```

```python
csgo_player.info(verbose=False)
```

#### CS:GO Players Results
Fill missing values in column 'ID' with 0.\
Change dtype column 'ID' from float to int.\
Drop 10 duplicate rows:
- Before 4399 
- After 4389

Extract CS:GO Players to Excel.

```python
csgo_player.to_excel(('csgo_player.xlsx'))
```

# 5. CS:GO Match Overview

```python
csgo_match_overview = pd.read_csv('matchOverview.csv')
csgo_match_overview.head()
```

Column 'MapName' need to change to 'MapID' from CS:GO Match Map.

```python
csgo_match_overview.info()
```

There is no missing value and column 'MatchID', 'Team1', 'Team2' dtype is already int.

```python
csgo_match_overview.duplicated().sum()
```

There is no duplicated value.

Adding column 'MapID' from CS:GO Match Map to CS:GO Match Overview by merging based on column 'MatchID' and 'MapName'.

```python
csgo_moverview_merge = csgo_match_overview.merge(csgo_match_map,left_on=['MatchID', 'MapName'], right_on=['MatchID', 'MapName'])
```

Placing column 'MapID' where column 'MapName' position is located by reindex the column.\
Also drop column 'MapName'.

```python
csgo_moverview_merge.columns
```

```python
csgo_moverview_merge = csgo_moverview_merge.reindex(['MatchID', 'MapName', 'MapID', 'Team1', 'Team2', 'Team1_Pick', 'Team1_Win',
                                                     'Team1_Final_Score', 'Team2_Final_Score', 'Halfs_Played',
                                                     'Team1_Start_Side', 'Team1_Half1_Score', 'Team2_Half1_Score',
                                                     'Team1_Half2_Score', 'Team2_Half2_Score', 'Team1_OT_Score',
                                                     'Team2_OT_Score'], axis=1)
csgo_moverview_merge.drop(columns='MapName', axis=1, inplace=True)
csgo_moverview_merge
```

```python
csgomap_map_unique = csgo_match_map['MapName'].nunique()
csgomap_mid_unique = csgo_match_map['MatchID'].nunique()

csgo_map_print = "CS:GO Match Map have "+str(csgomap_map_unique)+" unique 'MapName' values, and "+str(csgomap_mid_unique)+" unique 'MatchID' values"

csgoover_map_unique = csgo_match_overview['MapName'].nunique()
csgoover_mid_unique = csgo_match_overview['MatchID'].nunique()

csgo_match_overrow = len(csgo_match_overview)
csgo_mover_mergerow = len(csgo_moverview_merge)
lost_merge_row1 = (csgo_match_overrow-csgo_mover_mergerow)/csgo_match_overrow*100

print(csgo_map_print)
print("CS:GO Match Overview have "+str(csgoover_map_unique)+" unique 'MapName' values, and "+
      str(csgoover_mid_unique)+" unique 'MatchID' values")
print("Before, CS:GO Match Overview has "+str(csgo_match_overrow)+" rows. \
      \nCS:GO Match Overview now have "+str(csgo_mover_mergerow)+" rows.\
      \nCS:GO Match Overview lost "+str(round(lost_merge_row1, 2))+" rows after merge it with CS:GO Match Map based on column 'MatchID' and 'MapName'.")
```

#### CS:GO Match Overview Results
Merge CS:GO Match Overview with CS:GO Match Map based on column 'MatchID' and 'MapName' on both side
- Before 20016 rows
- After 1115 rows
- Lost 94.43 rows

Extract CS:GO Match Overview to Excel.

```python
csgo_moverview_merge.to_excel(('csgo_moverview_merge.xlsx'))
```
